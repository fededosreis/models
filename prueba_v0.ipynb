```
ventas["fecha"] = pd.to_datetime(ventas["fecha"])
envios["fecha"] = pd.to_datetime(envios["fecha"])
ventas["fecha"].head()
```

```
ventas['a√±o'] = pd.DatetimeIndex(ventas['fecha']).year
ventas['mes'] = pd.DatetimeIndex(ventas['fecha']).month
ventas.tail(10)
```

```
envios['a√±o'] = pd.DatetimeIndex(envios['fecha']).year
envios['mes'] = pd.DatetimeIndex(envios['fecha']).month
envios.tail(10)
```

```
ventas["canal"] = ventas["canal"].astype("category")
ventas["canal"].head()
```

```
ventas.isnull().sum()
```

```
envios.isnull().sum()
```

```
result = ventas.pivot_table(index='id_pos',
                      columns=['a√±o','mes'],
                      values='unidades', aggfunc=np.sum)
```

```
result.head(10)
```

```
ventas.groupby(['a√±o', 'canal'])['unidades'].aggregate('max').unstack()
```

```
ventas.groupby(['a√±o', 'mes'])['unidades'].aggregate('min').unstack()
```

```
res_1 = pd.Series.to_frame(result[2012,12])
```

```
res_1
```

```
res_2 = pd.Series.to_frame(result[2013,12])
```

```
ppp = [res_1,res_2]
res_anu=pd.concat(ppp)
```

```
res_anu
```

```
u = res_anu[2013,12].copy()
for p,i in enumerate(res_anu.index):
  if np.isnan(res_anu[2013,12].iloc[p]):
    if np.isnan(res_anu[2012,12].iloc[p]):
      u.iloc[p]=int(res_anu[2013,12].mean())
    else:
      u.iloc[p]=int(res_anu[2012,12].iloc[p])
  else:
    u.iloc[p]=int(res_anu[2013,12].iloc[p])
```

```
u = u.astype('int64')
```

```
u.to_csv('submission.csv')
```
